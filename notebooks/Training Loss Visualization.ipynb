```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc,rcParams
import numpy as np
import pickle

rc('axes', linewidth=2)
rc('font', weight='bold')
```

```python
path = '/Users/wslee-2/source/cs230/results/model_ae_25_2020-10-29-2102_dict'

results_dict = pickle.load(open(path,'rb'))

loss = results_dict['loss']
loss[-1]
```

```python
fig, ax = plt.subplots(1,2, figsize=[15,5])
ax[0].plot(np.arange(len(loss)),loss, linestyle='-', marker='o', color='b')
ax[0].scatter(10,loss[10])

# ax[0].text(11,0.95*loss[10],'loss = {:.2f}'.format(loss[10]),fontSize=10)
ax[0].set_ylabel('Training Loss', fontweight='bold')
ax[0].set_xlabel('Epoch', fontweight='bold')
ax[0].set_title('Training Loss', fontweight='bold')
ax[0].set_ylim([-0.1,1.21])

ax[1].plot(np.arange(len(loss)),loss, linestyle='-', marker='o', color='b')
ax[1].set_ylim([0,0.06])
ax[1].set_ylabel('Training Loss', fontweight='bold')
ax[1].set_xlabel('Epoch', fontweight='bold')
ax[1].set_title('Training Loss - Zoomed In', fontweight='bold')

plt.savefig(path.rsplit('.',1)[0] + '.png')
```
