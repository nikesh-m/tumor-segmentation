```python
filepath = '/Users/wslee-2/Downloads/brats-data/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/BraTS20_Training_001_flair.nii.gz'
epi_img = nib.load(filepath)
epi_img_data = epi_img.get_fdata()
epi_img_data.shape

```

```python
def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices),figsize=(20,10))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
rootdir = '/Users/wslee-2/Downloads/brats-data/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/'

for root, dirs, files in os.walk(rootdir):
    for name in files:
        filepath = os.path.join(root, name)
        print(name)

        slice_0 = epi_img_data[120, :, :]
        slice_1 = epi_img_data[:, 120, :]
        slice_2 = epi_img_data[:, :, 78]
        show_slices([slice_0, slice_1, slice_2])
        plt.suptitle("Center slices for EPI image") 
```
